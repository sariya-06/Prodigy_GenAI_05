```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image
```

```python
def load_img(path_to_img):
    max_dim = 512
    img = tf.io.read_file(path_to_img)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)

    shape = tf.cast(tf.shape(img)[:-1], tf.float32)
    long_dim = max(shape)
    scale = max_dim / long_dim

    new_shape = tf.cast(shape * scale, tf.int32)
    img = tf.image.resize(img, new_shape)
    img = img[tf.newaxis, :]
    return img
```

```python
print("Converting 'content.jpg' (AVIF) to 'content_converted.jpeg'...")
try:
    with PIL.Image.open("content.jpg") as img:
        img.save("content_converted.jpeg", "jpeg")
    content_image = load_img("content_converted.jpeg")
    print("'content.jpg' converted and loaded successfully.")
except Exception as e:
    print(f"Error converting or loading 'content.jpg': {e}")
    content_image = None

print("Converting 'style.jpg' (AVIF) to 'style_converted.jpeg'...")
try:
    with PIL.Image.open("style.jpg") as img:
        img.save("style_converted.jpeg", "jpeg")
    style_image = load_img("style_converted.jpeg")
    print("'style.jpg' converted and loaded successfully.")
except Exception as e:
    print(f"Error converting or loading 'style.jpg': {e}")
    style_image = None
```

```python
hub_model = hub.load(
'http://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2'
)
```

```python
stylized_image = hub_model(
tf.constant(content_image),
tf.constant(style_image)
)[0]
```

```python
import os

print(f"'content.jpg' exists: {os.path.exists('content.jpg')}")
print(f"'style.jpg' exists: {os.path.exists('style.jpg')}")
```

```python
def tensor_to_image(tensor):
    tensor = tensor * 255
    tensor = np.array(tensor, dtype=np.uint8)
    return PIL.Image.fromarray(tensor[0])
```

```python
tensor_to_image(stylized_image)
```
